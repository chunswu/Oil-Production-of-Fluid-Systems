```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv('data/dj_basin.csv')
print(df.describe())
```

```python
print(df.shape)
print(df.columns)
print(df.head()) 
print(df.info())
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(df['Prod30DayOil'], bins=100)
ax.set_xlabel('30 Day Production')
ax.set_ylabel('Number of Wells')
ax.set_title('30 Day Production')
plt.plot()
# df.hist('Prod30DayOil', figsize = [12, 6], bins=50)
```
