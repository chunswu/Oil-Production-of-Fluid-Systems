```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
df = pd.read_csv('data/dj_basin.csv')
print(df.describe())
```

```python
# print(df.shape)
# print(df.columns)
print(df.head(1)) 
# print(df.info())
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(df['Prod545DayOil'], bins=100)
ax.set_xlabel('Barrels', fontsize=24)
ax.set_ylabel('Number of Wells', fontsize=24)
ax.set_title('545 Day Oil Production', fontsize=34)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.savefig('images/eda_1.png')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(df['Norm545DayOil'], bins=100)
ax.set_xlabel('30 Day Production')
ax.set_ylabel('Number of Wells')
ax.set_title('Norm 30 Day Production')
plt.plot()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(df['Prod60DayOil'], bins=100)
ax.set_xlabel('30 Day Production (BBLS)')
ax.set_ylabel('Number of Wells')
ax.set_title('60 Day Production')
plt.plot()
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.hist(df['Norm30DayOil'], bins=100)
ax.set_xlabel('30 Day Production')
ax.set_ylabel('Number of Wells')
ax.set_title('Norm 30 Day Production')
plt.plot()
```
